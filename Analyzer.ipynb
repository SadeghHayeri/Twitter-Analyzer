```python
import tweepy
import tqdm
import numpy as np

from tqdm import tnrange, tqdm_notebook
```

```python
from secrets import consumer_key, consumer_secret, access_token, access_token_secret
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)
```

```python
def getUserFavorites(api, username, limit):
    res = np.array([])
    for favorite in tqdm_notebook(tweepy.Cursor(api.favorites, user=username).items(limit), unit="tweet", total=limit, desc=username):
        res = np.append(res, favorite)
    return res

def getAllFavorites(api, usersList, limit):
    res = {}
    for username in tqdm_notebook(usersList):
        res[username] = getUserFavorites(api, username, limit)
    return res
```
